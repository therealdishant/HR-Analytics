```python
# ---------------------------------------------
# üìÅ 1. Import Required Libraries
# ---------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
%matplotlib inline

```

```python
# ---------------------------------------------
# üìÇ 2. Load Dataset
# ---------------------------------------------
df = pd.read_csv(r"C:\Users\disha\Downloads\WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Quick overview
df.head()
```

```python
# ---------------------------------------------
# üìä 3. Basic Information
# ---------------------------------------------
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
df.info()
df.describe()
```

```python
# ---------------------------------------------
# üîç 4. Data Cleaning
# ---------------------------------------------
# Drop irrelevant columns
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True)

# Check for nulls
print(df.isnull().sum())

# Convert categorical columns to category dtype
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].astype('category')
```

```python
# ---------------------------------------------
# üìà 5. Univariate Analysis
# ---------------------------------------------

# Attrition distribution
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")

# Gender distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")

# Job Roles
plt.figure(figsize=(10, 4))
sns.countplot(y='JobRole', data=df)
plt.title("Job Role Distribution")
```

```python
# ---------------------------------------------
# üìä 6. Numerical Feature Distributions
# ---------------------------------------------
# Histograms
df.hist(bins=20, figsize=(18, 14), color='skyblue', edgecolor='black')
plt.suptitle("Numerical Feature Distributions", fontsize=20)
```

```python
# ---------------------------------------------
# üìä 7. Bivariate Analysis
# ---------------------------------------------
# Attrition vs Job Role
plt.figure(figsize=(12, 5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.title("Attrition by Job Role")

# Monthly Income vs Attrition
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")

# Overtime vs Attrition
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("OverTime vs Attrition")
```

```python
# ---------------------------------------------
# üìä 8. Correlation Heatmap
# ---------------------------------------------
# Encode categorical for correlation
df_encoded = pd.get_dummies(df, drop_first=True)

# Correlation matrix
plt.figure(figsize=(16, 12))
corr_matrix = df_encoded.corr()
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
```

```python
# ---------------------------------------------
# üìä 9. Feature Impact on Attrition (Custom Examples)
# ---------------------------------------------
# Satisfaction levels
satisfaction_cols = ['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']

for col in satisfaction_cols:
    sns.countplot(x=col, hue='Attrition', data=df)
    plt.title(f"{col} vs Attrition")
    plt.show()

# Years at company vs attrition
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df[df['Attrition'] == 'Yes'], x='YearsAtCompany', shade=True, label='Yes')
sns.kdeplot(data=df[df['Attrition'] == 'No'], x='YearsAtCompany', shade=True, label='No')
plt.title("Years at Company vs Attrition")
plt.legend()
```
